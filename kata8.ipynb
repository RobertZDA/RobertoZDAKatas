# Segundo problema

```python
planet_moons = {

    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1

}

moons = planet_moons.values()

planets = len(planet_moons)

print('Se tienen' , planets , 'planetas')

total_moons = 0

for luna in moons:

    total_moons = total_moons + luna

promedio = total_moons / planets

print('El promedio es:' , promedio)

```
