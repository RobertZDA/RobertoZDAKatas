# Primer problema

```python
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano' , 'Neptuno']

planetas.append('Pluton')

print('Existen',len(planetas),"planetas en nuestro sistema")

print('El ultimo planeta que agregue fue',planetas[-1])
```

# Segundo problema

```python
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano' , 'Neptuno']

eleccion = input("Busca el nombre de un planeta :D")

eleccion = eleccion.title()

eleccionindex = planetas.index(eleccion)

print("Los planetas mas cercanos son",planetas[0:eleccionindex])

print("Los planetas mas lejanos son ",planetas[eleccionindex + 1:])
```
